# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

#### A. Do you agree with the above analysis and recommendations? Why or why not?

The preliminary report seems to be based on visual inspection of the scatterplot, which shows readmissions clustered in the low discharge rate area, and a slight down slope if a linear regression fit were used. We will calculate the actual correlation coefficient.

```python
# Load the data into a pandas dataframe
# Avoid rows with NaN values (81 rows)
good_rows = clean_hospital_read_df['Excess Readmission Ratio'].notnull()
df_c = clean_hospital_read_df[good_rows]
discharges = df_c['Number of Discharges']
readmissions = df_c['Excess Readmission Ratio']
corr = np.corrcoef(discharges, readmissions)[0,1]
print('Pearson correlation coefficient: ', corr)
```

The small negative coefficient confirms the visual inspection. However, it cannot be said that there is a 'significant' correlation between the two. 

In addition, there in no basis to conclude from the given data that smaller hospitals lack the resources for quality patient care. Consequently, there is no reason to make the recommendationsfor smaller hospitals to upgrade.

#### B. Provide support for your arguments and your own recommendations with a statistically sound analysis:

We now set up our hypothesis test:
 - Null Hypothesis: discharges and readmissions are not correlated
 - Alternate Hypothesis: discharges and readmissions are correlated
 - Alpha = 0.01

```python
r, p = stats.pearsonr(discharges, readmissions)
print('Pearson correlation coefficient: ', r)
print('p-value: ', p)
```

The p-value being near zero, we can reject the null hypothesis, so the two variable are correlated, albeit slightly.
